```python
import torch
import torch.nn as nn
import torchvision
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

Start with instantiating a predefined torch vision model

```python
model = torchvision.models.vgg16(pretrained=False)
```

```python
model
```

```python
model.classifier
```

How to load a predefined model. For models trained on GPU but to be loaded on CPU, we need to add the map location and storage.

```python
torch.load('vgg16_torchvision_pretrain.pth',map_location=lambda storage, loc: storage)
```

As you can see we have now got the weights for the VGG16 layers

To see the architecture of the model, what layers the model contains we can use see the object of the model

```python
model
```

Now we can extract some of the layer features

```python
# Remove the last linear layer
del classifier[6]
classifier = nn.Sequential(*classifier)
```

To freeze few of the layers of the model we can set its requires_grad false.
This will ensure that the backprop wont calculate the gradients or they can be treated a constant/freezed

```python
for layer in features[:10]:
    for p in layer.parameters():
        p.requires_grad = False
```

```python
nn.Sequential(*features)
```
